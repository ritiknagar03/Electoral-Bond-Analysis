```python
def missing_Dtype_Unique(df):
  return pd.DataFrame({"Unique value": df.nunique(),
          "Data Type": df.dtypes,
          "Missing Value": df.isnull().sum()})
```

#Insert Data

```python
combined_df_company1['New Bonds No.']= combined_df_company1['Prefix']+combined_df_company1['Bond\rNumber'].astype(str)
combined_df_parties1['New Bonds No.']= combined_df_parties1['Prefix']+combined_df_parties1['Bond\rNumber'].astype(str)
```

#Remove Unneccesary Column

```python
combined_df_company1.drop(columns=['Sr No.', 'Reference No  (URN)', 'Journal Date', 'Date of Expiry', 'Prefix', 'Bond\rNumber','Issue Branch Code', 'Issue Teller'], inplace=True)
combined_df_parties1.drop(columns=['Sr No.', 'Account no. of\rPolitical Party', 'Prefix', 'Bond\rNumber', 'Pay Branch\rCode', 'Pay Teller'], inplace=True)
```

#Data Cleaning- column insert and type change

```python
combined_df_company1.columns
```

```python
combined_df_parties1.columns
```

```python
combined_df_company1['Date of\rPurchase']=pd.to_datetime(combined_df_company1['Date of\rPurchase'])
combined_df_parties1['Date of\rEncashment']=pd.to_datetime(combined_df_parties1['Date of\rEncashment'])

combined_df_company1['Denominations']=combined_df_company1['Denominations'].str.replace(',','').astype(int)
combined_df_parties1['Denominations']=combined_df_parties1['Denominations'].str.replace(',','').astype(int)

combined_df_company1['Purchase Day']=combined_df_company1['Date of\rPurchase'].dt.day_name()
combined_df_parties1['Encashment Day']=combined_df_parties1['Date of\rEncashment'].dt.day_name()
combined_df_company1['Purchase Month']=combined_df_company1['Date of\rPurchase'].dt.month
combined_df_parties1['Encashment Month']=combined_df_parties1['Date of\rEncashment'].dt.month
combined_df_company1['Purchase Year']=combined_df_company1['Date of\rPurchase'].dt.year
combined_df_parties1['Encashment Year']=combined_df_parties1['Date of\rEncashment'].dt.year
```

#Marge DF

```python
marge_df=pd.merge(combined_df_company1,combined_df_parties1,on='New Bonds No.', how='inner')
```
